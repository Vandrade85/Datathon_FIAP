## Funções

### FILTRA COLUNAS
Remove as colunas dos anos que passar como argumentos

```python
def filter_columns(df, filters: list): # adiciono no array o padrão que existe nas colunas e que não quero que tenha na saída final
    selected_columns = [True] * len(df.columns)  # Inicializa todas as colunas como True
    for index, column in enumerate(df.columns):
        if any(filter in column for filter in filters): selected_columns[index] = False
    return df[df.columns[selected_columns]]
```

## LIMPA DATASET
deleta todas as linhas onde há NA em todas as colunas, ou só há nome

```python
def cleaning_dataset(df):
  _df = df.dropna(subset=df.columns.difference(['NOME']), how='all') # executa o dropna para todas as colunas sem visualizar a coluna NOME
  _df = _df[~_df.isna().all(axis=1)] # remove linhas com apenas NaN, se tiver algum dado na linha não remove
  return _df
```

### Analise correlação
Faz uma análise de correlação do df

```python
def analyse_corr(df):
  import seaborn as sns
  import matplotlib.pyplot as plt

  df = df.apply(pd.to_numeric, errors='coerce')

  corr_matrix = df.corr()
  plt.figure(figsize=(10, 8))
  sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
  plt.show()
```

```python
def plot_evolucao_coluna(coluna, tipo_grafico='barra'):
    # Extrair os valores de cada DataFrame
    valores_2020 = df_2020[coluna + '_2020'].mean()  # Exemplo: calcular a média
    valores_2021 = df_2021[coluna + '_2021'].mean()
    valores_2022 = df_2022[coluna + '_2022'].mean()

    # Criar uma lista com os anos e os valores
    anos = ['2020', '2021', '2022']
    valores = [valores_2020, valores_2021, valores_2022]

    # Criar o gráfico conforme o tipo desejado
    plt.figure(figsize=(8, 5))  # Definir tamanho da figura

    if tipo_grafico == 'barra':
        plt.bar(anos, valores, color=['blue', 'orange', 'green'])
        plt.ylabel(coluna)
        for i, valor in enumerate(valores):
            plt.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')
    elif tipo_grafico == 'linha':
        plt.plot(anos, valores, marker='o', color='blue', linewidth=2)
        plt.ylabel(coluna)
        plt.ylim(3, max(valores) * 1.1)  # Define o eixo Y começando em 0, com um pouco de margem no topo

        for i, valor in enumerate(valores):
            plt.text(i, valor + (max(valores) * 0.02), f'{valor:.2f}', ha='center', va='bottom')  # Coloca as labels um pouco acima dos pontos
    elif tipo_grafico == 'pizza':
        plt.pie(valores, labels=anos, autopct='%1.1f%%', colors=['blue', 'orange', 'green'])
        plt.title(f'Participação de {coluna} ao longo dos anos')
    elif tipo_grafico == 'area':
        plt.fill_between(anos, valores, color='blue', alpha=0.5)
        plt.plot(anos, valores, marker='o', color='blue', linewidth=2)
        plt.ylabel(coluna)
    elif tipo_grafico == 'box':
        # Para o box plot, precisamos de uma lista de valores
        todos_valores = [df_2020[coluna + '_2020'], df_2021[coluna + '_2021'], df_2022[coluna + '_2022']]
        plt.boxplot(todos_valores, labels=anos)
        plt.ylabel(coluna)
    else:
        raise ValueError("Tipo de gráfico inválido. Use 'barra', 'linha', 'pizza', 'area' ou 'box'.")

    plt.xlabel('Ano')
    plt.title(f'Evolução de {coluna} ao longo dos anos')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
import matplotlib.pyplot as plt
import math
import seaborn as sns

def plot_evolucao_varias_colunas(colunas, tipo_grafico='barra'):
    num_colunas = len(colunas)
    max_colunas_por_linha = 4
    num_linhas = math.ceil(num_colunas / max_colunas_por_linha)  # Determinar quantas linhas serão necessárias

    fig, axs = plt.subplots(num_linhas, max_colunas_por_linha, figsize=(5 * max_colunas_por_linha, 5 * num_linhas))

    # Flatten axs se for necessário (para evitar problemas ao acessar os eixos)
    axs = axs.flatten() if num_colunas > 1 else [axs]

    for idx, coluna in enumerate(colunas):
        # Extrair os valores de cada DataFrame
        valores_2020 = df_2020[coluna + '_2020'].mean()  # Exemplo: calcular a média
        valores_2021 = df_2021[coluna + '_2021'].mean()
        valores_2022 = df_2022[coluna + '_2022'].mean()

        # Criar uma lista com os anos e os valores
        anos = ['2020', '2021', '2022']
        valores = [valores_2020, valores_2021, valores_2022]

        # Selecionar o subplot correspondente
        ax = axs[idx]

        # Criar o gráfico conforme o tipo desejado
        if tipo_grafico == 'barra':
            cores = sns.color_palette("Set2")  # Usando a paleta Set2 do seaborn
            ax.bar(anos, valores, color=cores[:3])  # Usando as 3 primeiras cores da paleta
            ax.set_ylabel(coluna)
            for i, valor in enumerate(valores):
                ax.text(i, valor, f'{valor:.2f}', ha='center', va='bottom')
        elif tipo_grafico == 'linha':
            ax.plot(anos, valores, marker='o', color='blue', linewidth=2)
            ax.set_ylabel(coluna)
            for i, valor in enumerate(valores):
                ax.text(i, valor, f'{valor:.2f}', ha='center', va='top')
        elif tipo_grafico == 'pizza':
            ax.pie(valores, labels=anos, autopct='%1.1f%%', colors=['blue', 'orange', 'green'])
            ax.set_title(f'Participação de {coluna} ao longo dos anos')
        elif tipo_grafico == 'area':
            ax.fill_between(anos, valores, color='blue', alpha=0.5)
            ax.plot(anos, valores, marker='o', color='blue', linewidth=2)
            ax.set_ylabel(coluna)
        elif tipo_grafico == 'box':
            todos_valores = [df_2020[coluna + '_2020'], df_2021[coluna + '_2021'], df_2022[coluna + '_2022']]
            ax.boxplot(todos_valores, labels=anos)
            ax.set_ylabel(coluna)
        else:
            raise ValueError("Tipo de gráfico inválido. Use 'barra', 'linha', 'pizza', 'area' ou 'box'.")

        ax.set_xlabel('Ano')
        ax.set_title(f'Evolução de {coluna} ao longo dos anos')
        ax.grid(axis='y', linestyle='--', alpha=0.7)

    # Remover eixos vazios se o número de gráficos for menor que o total de subplots
    for i in range(num_colunas, num_linhas * max_colunas_por_linha):
        fig.delaxes(axs[i])

    plt.tight_layout()  # Ajustar o layout para evitar sobreposição
    plt.show()

# Exemplo de chamada
colunas = ['IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN']
plot_evolucao_varias_colunas(colunas, tipo_grafico='barra')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import to_rgb
from unidecode import unidecode

# Função para padronizar os nomes das pedras
def padronizar_pedra(nome):
    return unidecode(nome).strip().upper()

# Listar os dataframes e rótulos correspondentes
df_list = [df_2020, df_2021, df_2022]
anos = ['2020', '2021', '2022']

# Padronizar os nomes das pedras em cada DataFrame
for i, df in enumerate(df_list):
    df[f'PEDRA_{anos[i]}'] = df[f'PEDRA_{anos[i]}'].apply(padronizar_pedra)

# Definir as cores base associadas às pedras
cores_pedras = {
    'QUARTZO': '#D3D3D3',   # Cinza claro
    'AGATA': '#FF4500',     # Vermelho/laranja
    'AMETISTA': '#8A2BE2',  # Roxo
    'TOPAZIO': '#FFFF00'    # Amarelo
}

# Função para gerar cores em degradê a partir da cor base
def gerar_degrade(cor_base, num_variacoes):
    cor = to_rgb(cor_base)
    return [(cor[0] * (1 - i / (num_variacoes)),
             cor[1] * (1 - i / (num_variacoes)),
             cor[2] * (1 - i / (num_variacoes))) for i in range(num_variacoes)]

# Contar a quantidade de cada pedra por ano e calcular o percentual
contagem_total = {}
total_ano = {}

for i, df in enumerate(df_list):
    contagem = df[f'PEDRA_{anos[i]}'].value_counts()
    total_ano[anos[i]] = contagem.sum()  # Total de pedras por ano
    for pedra in cores_pedras.keys():
        contagem_total[pedra] = contagem_total.get(pedra, [0, 0, 0])
        contagem_total[pedra][i] = contagem.get(pedra, 0) / total_ano[anos[i]] * 100  # Percentual

# Criar um DataFrame a partir do dicionário
contagem_df = pd.DataFrame.from_dict(contagem_total, orient='index', columns=anos)

# Criar o gráfico de barras
fig, ax = plt.subplots(figsize=(10, 6))

# Largura das barras
largura = 0.2

# Plotar barras com cores em degradê para cada tipo de pedra
for idx, pedra in enumerate(contagem_df.index):
    cor_degrade = gerar_degrade(cores_pedras[pedra], len(anos))
    for i in range(len(anos)):
        barra = ax.bar(idx + i * largura, contagem_df.loc[pedra][i], width=largura, color=cor_degrade[i])

        # Adicionar rótulo com o valor percentual em cima da barra
        ax.text(barra[0].get_x() + barra[0].get_width() / 2, barra[0].get_height(),
                f'{barra[0].get_height():.1f}%', ha='center', va='bottom')

# Adicionar rótulos dos anos no eixo Y
ax.set_title('Evolução Percentual das Pedras por Ano (2020, 2021, 2022)')
ax.set_xlabel('Tipo de Pedra')
ax.set_ylabel('Percentual (%)')
ax.set_xticks([x + largura for x in range(len(contagem_df.index))])
ax.set_xticklabels(contagem_df.index)
ax.set_ylim(0, 100)  # Percentual vai de 0 a 100%

# Ajustar o layout
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Função para separar os outliers
def separar_outliers(df, coluna):
    # Calcular os limites de 5%
    limite_inferior = df[coluna].quantile(0.05)
    limite_superior = df[coluna].quantile(0.95)

    # Criar DataFrames para os melhores e em desenvolvimento
    melhores = df[df[coluna] > limite_superior].sort_values(by=coluna, ascending=False).reset_index(drop=True)
    desenvolvimento = df[df[coluna] < limite_inferior].sort_values(by=coluna).reset_index(drop=True)

    return melhores, desenvolvimento

# Aplicar a função para cada DataFrame e coluna correspondente
melhores_2020, desenvolvimento_2020 = separar_outliers(df_2020, 'INDE_2020')
melhores_2021, desenvolvimento_2021 = separar_outliers(df_2021, 'INDE_2021')
melhores_2022, desenvolvimento_2022 = separar_outliers(df_2022, 'INDE_2022')
```

```python
# Função para remover o sufixo e juntar os DataFrames de melhores e desenvolvimento
def juntar_melhores_desenvolvimento(*dfs):
    # Remover sufixo de cada DataFrame e armazenar em uma lista
    dfs_sem_sufixo = [df.rename(columns=lambda x: x.replace('_2020', '').replace('_2021', '').replace('_2022', '')) for df in dfs]

    # Concatenar todos os DataFrames verticalmente
    df_final = pd.concat(dfs_sem_sufixo, ignore_index=True)

    return df_final

# Juntar os DataFrames de melhores
df_melhores = juntar_melhores_desenvolvimento(melhores_2020, melhores_2021, melhores_2022)

# Juntar os DataFrames de desenvolvimento
df_desenvolvimento = juntar_melhores_desenvolvimento(desenvolvimento_2020, desenvolvimento_2021, desenvolvimento_2022)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular a matriz de correlação para os melhores
correlacao_melhores = df_melhores[['INDE', 'IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN']].corr()

# Calcular a matriz de correlação para os desenvolvimento
correlacao_desenvolvimento = df_desenvolvimento[['INDE', 'IAA', 'IEG', 'IPS', 'IDA', 'IPP', 'IPV', 'IAN']].corr()

# Função para plotar a matriz de correlação triangular
def plotar_matriz_triangular(correlacao, titulo):
    mask = np.triu(np.ones_like(correlacao, dtype=bool))  # Criar máscara para a parte superior
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlacao, mask=mask, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
    plt.title(titulo)
    plt.show()

# Plotar as matrizes de correlação
plotar_matriz_triangular(correlacao_melhores, 'Matriz de Correlação - Melhores Alunos')
plotar_matriz_triangular(correlacao_desenvolvimento, 'Matriz de Correlação - Alunos em Desenvolvimento ')
```

```python
import pandas as pd
import re

# Função para normalizar os textos
def normalizar_destaque(texto):
    if pd.isna(texto):
        return None
    texto = texto.strip()
    # Normalizar para minúsculas
    texto = texto.lower()
    # Remover partes desnecessárias
    texto = re.sub(r'seu destaque em \d{4}: ', '', texto)
    texto = re.sub(r'ponto a melhorar em \d{4}: ', 'melhorar: ', texto)
    texto = re.sub(r'melhorar: ', '', texto)
    texto = re.sub(r'Destaque:', '', texto)

    # Categorizar notas boas
    if "boas notas" in texto:
        return "Notas Boas"
    elif "entrega das lições de casa" in texto:
        return "Entrega de Lições"
    elif "integração aos princípios" in texto:
        return "Integração"
    else:
        return texto

# Função para capitalizar a primeira letra
def capitalizar(texto):
    return texto.capitalize() if texto else texto

# Aplicar a função de normalização e capitalização
for coluna in ['DESTAQUE_IEG', 'DESTAQUE_IDA', 'DESTAQUE_IPV']:
    df_melhores[coluna] = df_melhores[coluna].apply(normalizar_destaque).apply(capitalizar)
    df_desenvolvimento[coluna] = df_desenvolvimento[coluna].apply(normalizar_destaque).apply(capitalizar)

# Verificar os valores únicos após a normalização e capitalização
print_unique_destaques(df_melhores, "Melhores (Normalizado)")
print_unique_destaques(df_desenvolvimento, "Desenvolvimento (Normalizado)")
```

```python
import matplotlib.pyplot as plt

# Função para plotar os subplots
def plotar_destaques_separados(df_melhores, df_desenvolvimento):
    fig, axs = plt.subplots(3, 1, figsize=(12, 15))

    # Melhores
    axs[0].barh(df_melhores['DESTAQUE_IEG'].value_counts().index, df_melhores['DESTAQUE_IEG'].value_counts(), color='lightblue', alpha=0.7)
    axs[0].set_title('Contagem de Destaque IEG - Melhores')
    axs[0].set_xlabel('Contagem')
    axs[0].set_ylabel('Destaques')

    axs[1].barh(df_melhores['DESTAQUE_IDA'].value_counts().index, df_melhores['DESTAQUE_IDA'].value_counts(), color='lightblue', alpha=0.7)
    axs[1].set_title('Contagem de Destaque IDA - Melhores')
    axs[1].set_xlabel('Contagem')
    axs[1].set_ylabel('Destaques')

    axs[2].barh(df_melhores['DESTAQUE_IPV'].value_counts().index, df_melhores['DESTAQUE_IPV'].value_counts(), color='lightblue', alpha=0.7)
    axs[2].set_title('Contagem de Destaque IPV - Melhores')
    axs[2].set_xlabel('Contagem')
    axs[2].set_ylabel('Destaques')

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajustar o layout
    plt.suptitle('Melhores', fontsize=16)
    plt.show()

    # Desenvolvimento
    fig, axs = plt.subplots(3, 1, figsize=(12, 15))

    axs[0].barh(df_desenvolvimento['DESTAQUE_IEG'].value_counts().index, df_desenvolvimento['DESTAQUE_IEG'].value_counts(), color='lightcoral', alpha=0.7)
    axs[0].set_title('Contagem de Destaque IEG - Desenvolvimento')
    axs[0].set_xlabel('Contagem')
    axs[0].set_ylabel('Destaques')

    axs[1].barh(df_desenvolvimento['DESTAQUE_IDA'].value_counts().index, df_desenvolvimento['DESTAQUE_IDA'].value_counts(), color='lightcoral', alpha=0.7)
    axs[1].set_title('Contagem de Destaque IDA - Desenvolvimento')
    axs[1].set_xlabel('Contagem')
    axs[1].set_ylabel('Destaques')

    axs[2].barh(df_desenvolvimento['DESTAQUE_IPV'].value_counts().index, df_desenvolvimento['DESTAQUE_IPV'].value_counts(), color='lightcoral', alpha=0.7)
    axs[2].set_title('Contagem de Destaque IPV - Desenvolvimento')
    axs[2].set_xlabel('Contagem')
    axs[2].set_ylabel('Destaques')

    plt.tight_layout(rect=[0, 0, 1, 0.95])  # Ajustar o layout
    plt.suptitle('Desenvolvimento', fontsize=16)
    plt.show()

# Chamar a função
plotar_destaques_separados(df_melhores, df_desenvolvimento)
```

```python
# Criar um dataframe separado para ingressantes e veteranos
ingressantes = media_por_categoria_e_fase[media_por_categoria_e_fase['SINALIZADOR_INGRESSANTE_2021'] == 'Ingressante']
veteranos = media_por_categoria_e_fase[media_por_categoria_e_fase['SINALIZADOR_INGRESSANTE_2021'] == 'Veterano']

# Juntar os dados de ingressantes e veteranos por FASE_2021
df_comparacao = pd.merge(ingressantes, veteranos, on='FASE_2021', suffixes=('_ingressante', '_veterano'))

# Calcular a diferença percentual por fase
df_comparacao['diferenca_percentual'] = ((df_comparacao['INDE_2021_veterano'] - df_comparacao['INDE_2021_ingressante']) / df_comparacao['INDE_2021_ingressante']) * 100

# Verificando a tabela de comparação
print(df_comparacao)
```

```python
# Configurando o gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(10, 6))

# Criando o gráfico de barras com Ingressantes e Veteranos separados por FASE_2021
sns.barplot(x='FASE_2021', y='INDE_2021', hue='SINALIZADOR_INGRESSANTE_2021', data=media_por_categoria_e_fase, ax=ax)

# Adicionando a diferença percentual no gráfico para cada fase
for i, row in df_comparacao.iterrows():
    # Posição horizontal ajustada para o centro entre as barras de ingressante e veterano
    x_pos = i
    y_pos = max(row['INDE_2021_ingressante'], row['INDE_2021_veterano']) + 0.2  # Posição do texto acima da barra mais alta
    ax.text(x_pos, y_pos, f'{row["diferenca_percentual"]:.2f}%', ha='center', fontsize=10, color='black')

# Ajustando o título e rótulos
ax.set_title('Comparação de Média de INDE por Fase entre Ingressantes e Veteranos', pad=20)
ax.set_ylabel('Média de INDE')
ax.set_xlabel('Fase (Ano)')
# Ajustando o limite superior do eixo y para 10
ax.set_ylim(0, 10)

# Exibir o gráfico
plt.show()
```
